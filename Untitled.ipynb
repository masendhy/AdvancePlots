```python
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
df.head()
```

## Subplots :
 * Digunakan untuk membuat plot dalam beberapa axes. plt.subplot() memiliki dua output yaitu figure dan axes. Serta memiliki input berupa jumlah axes yang ingin dibuat (baris dan kolom)

```python
fig, ax = plt.subplots(1,1, figsize=(8,5)) # 1,1 artinya ; jumlah row nya adh 1 dan jumlah kolomnya adalah 1
ax.plot(df['Year'],df['Physical Sciences'],color='red' )
```

Axes memiliki bentuk seperti matriks, maka untuk memilih axes tersebut dapat memamnggil seperti matriks.<br>
    * ax[0] = untuk axes yang berada di titik [0], jika hanya terdapat 1 baris 1 kolom<br>
    * ax[0,1] = untuk axes yang berada dibaris 0 dan row ke 1<br>

```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5))

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')
```

* Tambahkan **linestyle** untuk mengubah linestyle nya

1. Solid (‘-‘) – jenis garis penuh dan menjadi default ketika tidak ada jenis linestyle yang ditentukan.

2. Dashed (‘–-‘) – jenis garis putus-putus yang terdiri dari beberapa bagian dengan jarak yang sama.

3. Dash-dot (‘-.’) – jenis garis yang terdiri dari urutan dash-dot.

4. Dotted (‘:’) – jenis garis yang terdiri dari titik-titik kecil.

5. None – jenis linestyle yang tidak memiliki garis atau dapat dibilang tidak terlihat.


```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red' ,linestyle=('-.'))
```

* Menggunakan **marker** untuk memperindah plot grafik yang dibuat

1. ‘.’ – titik kecil
2. ‘o’ – lingkaran
3. ‘v’ – segitiga ke bawah
4. ‘^’ – segitiga ke atas
5. ‘<‘ – segitiga ke kiri
6. ‘>’ – segitiga ke kanan
7. ‘s’ – persegi
8. ‘p’ – segi lima
9. ‘*’ – bintang
10. ‘h’ – heksagon
11. ‘+’ – simbol plus
12. ‘x’ – simbol x
13. ‘D’ – diamond
14. ‘d’ – diamond kecil
15. ‘_’ – garis horizontal

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red' ,linestyle=('-'), marker='o',markerfacecolor='blue')
```

* Mengubah **judul** dengan menggunakan matplotlib

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red' )

plt.title('Physical Science Bachelor Over Year',fontsize=15)
```

* Menggunakan **set_title** untuk memberikan judul pada matriks yang memiliki axes lebih dari satu
* dan menggunakan **fig.tight_layout** untuk memberikan jarak antar matriks axes nya

```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5))

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')

ax[0].set_title('Physical Sciences Bachelors Over Year',fontsize='15')
ax[1].set_title('Computer Sciences Bachelors Over Year',fontsize='15')

fig.tight_layout(pad=3.9)
```

* Memberikan **label** pada axis matriks

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red')

plt.xlabel('Year',fontsize='15')
plt.ylabel('Percentage Bachelor', fontsize='15')
```

* Memberikan **label** pada matriks yang memiliki lebih dari 1 axis

```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5))

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')

ax[0].set_title('Physical Sciences Bachelors Over Year',fontsize='15')
ax[1].set_title('Computer Sciences Bachelors Over Year',fontsize='15')

ax[1].set_xlabel('Year',fontsize='13')
ax[0].set_ylabel('Percentage Bachelor', fontsize='13')
ax[1].set_ylabel('Percentage Bachelor', fontsize='13')

fig.tight_layout(pad=3.0)
```
