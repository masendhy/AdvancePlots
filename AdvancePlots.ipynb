```python
import pandas as pd
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
df.head()
```

## Subplots :
 * Digunakan untuk membuat plot dalam beberapa axes. plt.subplot() memiliki dua output yaitu figure dan axes. Serta memiliki input berupa jumlah axes yang ingin dibuat (baris dan kolom)

```python
fig, ax = plt.subplots(1,1, figsize=(8,5)) # 1,1 artinya ; jumlah row nya adh 1 dan jumlah kolomnya adalah 1
ax.plot(df['Year'],df['Physical Sciences'],color='red' )
```

Axes memiliki bentuk seperti matriks, maka untuk memilih axis tersebut dapat memamnggil seperti matriks.<br>
    * ax[0] = untuk axes yang berada di titik [0], jika hanya terdapat 1 baris 1 kolom<br>
    * ax[0,1] = untuk axes yang berada dibaris 0 dan row ke 1<br>

```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5))

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')
```

* Tambahkan **linestyle** untuk mengubah linestyle nya

1. Solid (‘-‘) – jenis garis penuh dan menjadi default ketika tidak ada jenis linestyle yang ditentukan.

2. Dashed (‘–-‘) – jenis garis putus-putus yang terdiri dari beberapa bagian dengan jarak yang sama.

3. Dash-dot (‘-.’) – jenis garis yang terdiri dari urutan dash-dot.

4. Dotted (‘:’) – jenis garis yang terdiri dari titik-titik kecil.

5. None – jenis linestyle yang tidak memiliki garis atau dapat dibilang tidak terlihat.


```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red' ,linestyle=('-.'))
```

* Menggunakan **marker** untuk memperindah plot grafik yang dibuat

1. ‘.’ – titik kecil
2. ‘o’ – lingkaran
3. ‘v’ – segitiga ke bawah
4. ‘^’ – segitiga ke atas
5. ‘<‘ – segitiga ke kiri
6. ‘>’ – segitiga ke kanan
7. ‘s’ – persegi
8. ‘p’ – segi lima
9. ‘*’ – bintang
10. ‘h’ – heksagon
11. ‘+’ – simbol plus
12. ‘x’ – simbol x
13. ‘D’ – diamond
14. ‘d’ – diamond kecil
15. ‘_’ – garis horizontal

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red' ,linestyle=('-'), marker='o',markerfacecolor='blue')
```

* Mengubah **judul** dengan menggunakan matplotlib

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red' )

plt.title('Physical Science Bachelor Over Year',fontsize=15)
```

* Menggunakan **set_title** untuk memberikan judul pada matriks yang memiliki plot lebih dari satu (memiliki subplot)
* dan menggunakan **fig.tight_layout** untuk memberikan jarak antar matriks plot nya

```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5))

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')

ax[0].set_title('Physical Sciences Bachelors Over Year',fontsize='15')
ax[1].set_title('Computer Sciences Bachelors Over Year',fontsize='15')

fig.tight_layout(pad=3.9)
```

* Memberikan **label** pada axis matriks

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red')

plt.xlabel('Year',fontsize='15')
plt.ylabel('Percentage Bachelor', fontsize='15')
```

* Memberikan **label** pada matriks yang memiliki lebih subplots

```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5))

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')

ax[0].set_title('Physical Sciences Bachelors Over Year',fontsize='15')
ax[1].set_title('Computer Sciences Bachelors Over Year',fontsize='15')

ax[1].set_xlabel('Year',fontsize='13')
ax[0].set_ylabel('Percentage Bachelor', fontsize='13')
ax[1].set_ylabel('Percentage Bachelor', fontsize='13')

fig.tight_layout(pad=3.0)
```

* Melakukan **limitasi** pada matriks

```python
fig, ax = plt.subplots(1,1, figsize=(8,5))
ax.plot(df['Year'],df['Physical Sciences'],color='red')

plt.title('Physical Sciences Over Year', fontsize='15')

plt.xlim(1990,2010)
plt.ylim(0,50)
```

* Melakukan **limitasi** pada matriks yang memiliki subplots, jangan lupa untuk melakukan set sharex dan sharey menjadi True<br>
* Menggunakan **set_xticks** untuk mengubah range pada axis tertentu


```python
fig, ax =  plt.subplots(2, # number of rows
                        1, # number of columns
                       figsize=(8,5),
                       sharex=True,
                       sharey=True)

ax[0].plot(df['Year'],df['Physical Sciences'],color='red')
ax[1].plot(df['Year'],df['Computer Science'],color='blue')

ax[0].set_xlim(1990,2010)
ax[1].set_xlim(2000,2010)

ax[0].set_ylim(0,50)
ax[1].set_ylim(0,50)

ax[0].set_xticks(range(1990,2014,4)) # mengubah range pada axis x dari tahun 1990 s/d tahun 2014 dengan kelipatan 4
```

* Menambahkan **legenda** untuk membedakan beberapa data yang terdapat dalam 1 plot<br>
* gunakan **loc** untuk menentukan posisi legendanya

```python
plt.figure(figsize=(8,5))

plt.plot(df['Year'],df['Physical Sciences'],color='blue',label='Physical Science')
plt.plot(df['Year'],df['Computer Science'],color='red',label='Computer Science')

plt.legend(loc='lower right')
```

* Menambahkan **text** pada matriks

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Physical Sciences'], color ='blue')
plt.plot(df['Year'], df['Computer Science'], color ='green')

plt.text(1975, 25, s ='computer science',
         fontsize=11,
         color='green',
         rotation=63
        ) # memberikan text "computer science" pada sumbu x = 1975 dan sumbu y = 25
```

* Menambahkan **anotasi**, yaitu digunakan unutk menunjukkan / memberi keterangan pada hal penting.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'],df['Computer Science'],color='red')
plt.ylim(10,50)

plt.annotate('maximum', xy=(1983,37), xytext=(1983+5, 37+5),fontsize=15,
            arrowprops=dict(ec='black',arrowstyle='simple'))
```

* Menambahkan **horizontal line** untuk menunjukkan *threshold* (ambang batas) dari nilai terterntu

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Physical Sciences'], color='red')

plt.annotate('threshold', xy=(1970,26))
plt.axhline(y=25, color='green', ls='--',lw='1.5')
```

* Menambahkan **vertical line** untuk menunjukkan suatu nilai yang akan di highlight pada plot

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'],df['Physical Sciences'],color='blue')

plt.annotate('important event',xy=(1985+1,15))
plt.axvline(x=1985,color='red')
```
